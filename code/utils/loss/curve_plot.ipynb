```python
import csv
import matplotlib.pyplot as plt

# 读取CSV文件
with open('/home/wangye/YeProject/elite_experiments/global_mapping_20230727_896_384_res/csv.csv', 'r') as csvfile:
   reader = csv.reader(csvfile)
   data = [row for row in reader]

# 提取时间点和损失值
time = [float(row[0]) for row in data]
loss = [float(row[1]) for row in data]

# 绘制损失曲线图
plt.plot(time, loss)
plt.xlabel('Time')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.show()
```
